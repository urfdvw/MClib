# Experiments on the measure

## chi-square like measure
$$\begin{array}{l}
D_{\chi^2 } =\int \frac{\pi^{\prime } {\left(x\right)}^2 }{\pi \left(x\right)}\;\mathit{dx}-1\\
=\int \frac{\sum_m {\left\lbrack w^{\left(m\right)} \delta \left(x-x^{\left(m\right)} \right)\right\rbrack }^2 }{\pi \left(x\right)}\;\mathit{dx}-1
\end{array}$$
because
$$\frac{\sum_m {\left\lbrack w^{\left(m\right)} \delta \left(x-x^{\left(m\right)} \right)\right\rbrack }^2 }{\pi \left(x\right)}=\left\lbrace \begin{array}{cc}
0 & x\not= x^{\left(m\right)} \\
\frac{{w^{\left(m\right)} }^2 {\delta \left(0\right)}^2 }{\pi \left(x^{\left(m\right)} \right)} & x=x^{\left(m\right)} 
\end{array}\right.$$
Therefore
$$D_{\chi^2 } =\sum_m \frac{{w^{\left(m\right)} }^2 {\delta \left(0\right)}^2 }{\pi \left(x^{\left(m\right)} \right)}-1$$
TO remove the infinity
$$\tilde{D} =\frac{D_{\chi^2 } +1}{{\delta \left(0\right)}^2 }=\sum_m \frac{{w^{\left(m\right)} }^2 }{\pi \left(x^{\left(m\right)} \right)}$$

## Target distribution

$$ \pi(x) = \mathcal{N}(\mathbf{0},I) $$

```python
def pi(x): return mvn.pdf(x, mean=np.zeros(shape=note.D), cov=np.eye(note.D))
```

## normalized distance
With one modification, the measure is much less sensitive to the size of the samples.

$$\tilde{D}_N =M\sum_m \frac{{w^{\left(m\right)} }^2 }{\pi \left(x^{\left(m\right)} \right)}$$

Histogram of log measure when number of samples is 1000

```python
N = 1000
note.hist(note.samplesize, sq.Dchi2N, pi, N)
```

* * *
Histogram of log measure when number of samples is 10000

```python
N = 10000
note.hist(note.samplesize, sq.Dchi2N, pi, N)
```

This measure is much less sensitive to sample size.

The normalized measure is used in the rest of the experiments.

## Sample from different distributions and measure the distance

The y-axis is always log measure, and x-axis is always $\tau$.

$$
x\sim N\left(\left\lbrack \begin{array}{c}
\tau \\
0\\
\vdots \\
0
\end{array}\right\rbrack ,\mathit{\mathbf{I}}\right)
$$

```python
tau = np.linspace(-20.0,20.0,500)
note.plot(note.sampleshift, sq.Dchi2N, pi, tau)
```

The true position, $\tau=0$, has the smallest distance

* * *
$$
\begin{array}{l}
c \sim \mathrm{binomial}\left(\tau \right)\\
\mathrm{if}\;c==1\\
\;\;\;\;x_{i\;} \overset{\mathrm{iid}}{\sim} \;\mathrm{uniform}\left(-3,3\right)\\
\mathrm{else}\\
\;\;\;\;x\sim \pi \left(x\right)
\end{array}
$$

```python
tau = np.linspace(0.0,1.0,500)
note.plot(note.sampleshape, sq.Dchi2N, pi, tau)
```

$\tau=0$ gives the true target distribution,
where the distance is the smallest.

***
$$
x \sim \mathcal{N}(\mathbf{0},\ \tau \cdot \mathbf{I})
$$

```python
tau = np.linspace(0.0001,5.0,500)
note.plot(note.samplescale, sq.Dchi2N, pi, tau)
```

True scaling, $\tau = 1$ does not have the smallest distance.
The distance is minimized when all samples are at the peak of the target distribution,
i.e. $x\sim\mathcal{N}(\mathbf{0},\mathbf{0})$.

Let's zoom in.

```python
tau = np.linspace(0.0001,1.2,500)
note.plot(note.samplescale, sq.Dchi2N, pi, tau)
```

Instead of matching the distribution, the proposed distance prefers a single location of local maximum.
